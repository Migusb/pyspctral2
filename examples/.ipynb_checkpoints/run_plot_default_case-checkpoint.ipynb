## Run and plot for default case

```python
m = psp2.model()  # no args => default settings (based on those found in the original C run program)
#m = psp2.model(tau500=0.4, watvap=2.0, ozone=-1.0)
print('lat, lon:', m.lat, m.lon)
print('datetime:', m.dt)
print('temp, press:', m.temp, m.press)
print('tau500, watvap, ozone:', m.tau500, m.watvap, m.ozone)
```

```python
m.run()
```

```python
wl, Idr, Idf = np.loadtxt(m.raw_ofname, delimiter=',', skiprows=1, unpack=True)
f1, a = plt.subplots(figsize=(8, 4))
a.plot(wl, Idr, label='direct')
a.plot(wl, Idf, label='diffuse')
a.plot(wl, Idr+Idf, label='total')
a.set_xlim((0.3, 2.6))
a.set_xlabel('wavelength (Î¼m)')
a.set_ylabel('spectral irradiance (W m$^{-2}$ Î¼m$^{-1}$)')
a.legend()
a.grid(True)
f1.tight_layout();
```
